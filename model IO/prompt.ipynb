# 1. 导入配置信息

```python
from dotenv import find_dotenv
import dotenv

# 加载环境变量
dotenv.load_dotenv(find_dotenv(filename=".env"))

import os
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["OPENAI_API_BASE"] = os.getenv("OPENAI_BASE_URL")
os.environ["OPENAI_MAX_TOKENS"] = os.getenv("OPENAI_MAX_TOKENS")
os.environ["OPENAI_MODEL"] = os.getenv("OPENAI_MODEL")
os.environ["OPENAI_TEMPERATURE"] = os.getenv("OPENAI_TEMPERATURE")
os.environ["OPENAI_TOP_P"] = os.getenv("OPENAI_TOP_P")

print(os.environ["OPENAI_API_KEY"])
print(os.environ["OPENAI_API_BASE"])
```

# 2. 创建 ChatClient

```python
from langchain_openai import ChatOpenAI

chat_model = ChatOpenAI(
    model_name=os.getenv("OPENAI_MODEL"),
    base_url=os.getenv("OPENAI_API_BASE"),
    api_key=os.getenv("OPENAI_API_KEY"),
    temperature=float(os.getenv("OPENAI_TEMPERATURE")),
    max_tokens=int(os.getenv("OPENAI_MAX_TOKENS")),
    top_p=float(os.getenv("OPENAI_TOP_P")),
)
```

# 3. 创建 PromptTemplate
核心方法: `from_template(template: str)`

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template(
    template="请简单介绍一下{topic}?",
    partial_variables={"topic": "长沙美食"},  # 默认值
)

for chunk in chat_model.stream(prompt_template.format(topic="广东美食")):
    print(chunk.content, end="", flush=True)
```

# 4. 创建 ChatPromptTemplate
核心方法

`from_messages(messages: List[BaseMessage])`: 创建一个 ChatPromptTemplate 对象

`SystemMessagePromptTemplate`: 创建一个系统消息模板

`HumanMessagePromptTemplate`: 创建一个用户消息模板


```python
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

chat_prompt_template = ChatPromptTemplate.from_messages(
    messages=[
        SystemMessagePromptTemplate.from_template("你是一个只会阿巴阿巴的智障机器人"),
        HumanMessagePromptTemplate.from_template("请简单介绍一下{topic}?"),
    ],
)

result = chat_prompt_template.invoke({"topic": "长沙"})
print(result)
```
